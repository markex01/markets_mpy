```python
import os
from moviepy.video.io.VideoFileClip import VideoFileClip
os.listdir()

video_path = os.path.abspath("../data/video")
audio_path = os.path.abspath("../data/audio")
print(video_path)

for file in os.listdir(video_path):
    if file.endswith(".mp4"):
        # check if the file has already been converted to mp3
        mp3_file = file.replace(".mp4", ".mp3")
        if not mp3_file in os.listdir(audio_path):
            print(f"Converting {file} to {mp3_file}...")
            video = VideoFileClip(os.path.join(video_path, file))
            video.audio.write_audiofile(os.path.join(audio_path, mp3_file))
            video.close()
            print(f"Converted {file} to {mp3_file}.")
        else:
            print(f"{mp3_file} already exists. Skipping conversion.")
```

```python
import pdfplumber
import pandas as pd

pdf_path = 'Graduatoria_DM_FERXtransitorio_PC_FTV_2025_TabA.pdf'

# PDF page numbers you care about
start_pdf_page = 1
end_pdf_page   = 10   # inclusive

# convert to 0-index for pdfplumber
start_idx = start_pdf_page - 1
end_idx   = end_pdf_page - 1

dfs = []

with pdfplumber.open(pdf_path) as pdf:
    for page_idx in range(start_idx, end_idx + 1):
        page = pdf.pages[page_idx]

        # Try to detect the big table on the page
        table = page.extract_table(table_settings={
            "vertical_strategy": "lines",
            "horizontal_strategy": "lines",
            "intersection_tolerance": 5,
        })

        if table is None:
            print(f"No table found on page index {page_idx}")
            continue

        # table is a list of rows -> first row usually header
        header, *rows = table
        df = pd.DataFrame(rows, columns=header)
        dfs.append(df)

# Merge pages into a single DataFrame
if dfs:
    full_table = pd.concat(dfs, ignore_index=True)
    print(full_table.head())
else:
    print("No tables found in the requested pages")
```

```python
import geopandas as gpd

df_dcs_us = df_dcs[df_dcs['Market'] == 'US'].copy()

df_dcs_us['Date'] = pd.to_datetime(df_dcs_us['Date'])
df_dcs_us.sort_values(by='Date', ascending=False, inplace=True)
df_dcs_us = df_dcs_us[df_dcs_us['Date'] >= '2025-08-31']

# Load US states shapefile
states = gpd.read_file("https://www2.census.gov/geo/tiger/GENZ2022/shp/cb_2022_us_state_20m.zip")

# Convert to GeoDataFrame
gdf = gpd.GeoDataFrame(
    df_dcs_us,
    geometry=gpd.points_from_xy(df_dcs_us['Longitude'], df_dcs_us['Latitude']),
    crs="EPSG:4326"
)

# Spatial join
result = gpd.sjoin(gdf, states[['NAME', 'geometry']], how='left', predicate='within')
df_dcs_us['State'] = result['NAME'].values
```
