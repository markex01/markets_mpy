```python
qe = QuarterlyDataExtractor(
    Extractor=BE,
    quarter=quarterly,
    europe_daily_pool_securities=europe_daily_pool_securities,
    europe_hourly_pool_security_roots=europe_hourly_pool_security_roots,
    power_prices_path=europe_daily_pool_path
)
```

```python
# create a loop for all the securities and export them to excel
start_date = "2024-01-01"
for value in exports_abs.values():
    qe.build_monthly_absolute_df(securities_dict=value["securities"], start_date=start_date, export_path=value["excel_path"], periodicity=value["periodicity"])
```

```python
start_date = "2024-01-01"
for index, value in exports_pu.items():
    qe.build_monthly_normalized_df(securities_dict=value["securities"], start_date=start_date, export_path=value["excel_path"], periodicity=value["periodicity"])
```

```python
fx_forecast_securities = {
    "EURUSD": [
        "FCUSEU Q426 Index",
        "FCUSEU Q427 Index",
        "FCUSEU Q428 Index",
        "FCUSEU Q429 Index",
    ],
    "EURGBP": [
        "FCEUGB Q426 Index",
        "FCEUGB Q427 Index",
        "FCEUGB Q428 Index",
        "FCEUGB Q429 Index",
    ],
    "EURPLN": [
        "FCEUPL Q426 Index",
        "FCEUPL Q427 Index",
        "FCEUPL Q428 Index",
        "FCEUPL Q429 Index",
    ],
}

all_dfs = []
for key, securities_list in fx_forecast_securities.items():
    # Convert list to dict: strip " Index" suffix to use as column identifier
    securities_dict = {s.replace(" Index", f" {key}"): s for s in securities_list}
    df = qe.build_monthly_absolute_df(securities_dict=securities_dict, periodicity="M", start_date="2026-01-01", end_date="2026-01-31")
    all_dfs.append(df.T.set_index("date"))  # un-transpose; use date as index

df_all = pd.concat(all_dfs, axis=1)  # join all currency pairs as columns
df_all.to_excel(fx_forecasts_path)
df_all
```
